## Test for Text summarization from the json file

```python
import json

# Load the data
with open('test_subtitles_th.json', 'r') as f:
    data = json.load(f)

# Join all 'text' values into one string
full_text = " ".join([item['text'] for item in data])

print(full_text)
```

```python
import re
from collections import Counter

def extract_key_sentences(content, n=1):
    # 1. Clean and split into sentences
    sentences = re.split(r'(?<=[.!?]) +', content)
    
    # 2. Find most common words (ignoring short grammar words)
    words = re.findall(r'\w+', content.lower())
    stop_words = {'i', 'me', 'my', 'to', 'the', 'a', 'and', 'is', 'in', 'it', 'of', 'that', 'you', 'we', 're'}
    meaningful_words = [w for w in words if w not in stop_words and len(w) > 2]
    word_freq = Counter(meaningful_words)
    
    # 3. Score sentences based on word frequency
    scores = {}
    for sentence in sentences:
        for word in re.findall(r'\w+', sentence.lower()):
            if word in word_freq:
                scores[sentence] = scores.get(sentence, 0) + word_freq[word]
    
    # 4. Return top N sentences
    return sorted(scores, key=scores.get, reverse=True)[:n]

# Extract the most important sentence
important_part = extract_key_sentences(full_text, 1)
print(f"Extracted: {important_part[0]}")
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from langdetect import detect
# 1. Detect language (or specify it if known, e.g., 'english', 'french', 'chinese')
# For a "catch-all," English works surprisingly well for many Latin-based languages
text = "สวัสดีครับ ยินดีต้อนรับสู่รายการของเรา วันนี้ผมจะมาพูดถึงวิธีการออมเงินอย่างง่ายสำหรับวัยทำงานครับ ปัญหาหลักของหลายคนคือการใช้เงินแบบเดือนชนเดือน ดังนั้นสิ่งแรกที่เราต้องทำคือการบันทึกรายรับรายจ่ายครับ การจดบันทึกจะทำให้เราเห็นว่าเงินหายไปกับค่าอะไรที่ไม่จำเป็นบ้าง เช่น ค่ากาแฟหรือค่าขนมจุกจิก ต่อมาคือกฎการออมเงิน 20 เปอร์เซ็นต์ เมื่อเงินเดือนออก ให้เราหักเงินออมไว้ก่อนเลยครับ อย่ารอให้เหลือแล้วค่อยออม เพราะสุดท้ายแล้วมันจะไม่มีเหลือ สุดท้ายคือการลงทุนครับ การเก็บเงินไว้ในธนาคารเฉยๆ อาจจะไม่เพียงพอเนื่องจากเงินเฟ้อ เราควรศึกษาเรื่องกองทุนรวมหรือหุ้นพื้นฐานเพื่อเพิ่มพูนความมั่งคั่งในระยะยาวครับ"
language = detect(full_text)


# 2. Parse the text
parser = PlaintextParser.from_string(full_text, Tokenizer(language))

# 3. Use LSA Summarizer
summarizer = LsaSummarizer()

# 4. Extract the top 2 most important sentences
summary = summarizer(parser.document, 2)

for sentence in summary:
    print(sentence)
```
